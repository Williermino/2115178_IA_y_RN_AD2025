# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label = "Seno de x")  # Dibuja el gráfico
plt.title("Gráfico de Seno")  # Título
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("sin(x)")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```

```python
x = np.random.rand(100)  # Genera 100 puntos aleatorios en x
y = np.random.rand(100)  # Genera 100 puntos aleatorios en y

plt.scatter(x, y, color='blue', label="Datos Aleatorios")  # Dibuja el gráfico de dispersión
plt.title("Gráfico de Dispersión")  # Título
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("y")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # se puede ver como una tabla, es decir, 1 fila con 3 columnas

axs[0].plot(x, y1, label="Seno de x")
axs[0].set_title("Gráfico de Seno")
axs[0].legend()

axs[1].plot(x, y2, label="Coseno de x", color='red')
axs[1].set_title("Gráfico de Coseno")
axs[1].legend()

axs[2].plot(x, y2, label="Coseno de x", color='orange')
axs[2].set_title("Gráfico de Coseno")
axs[2].legend()

plt.tight_layout()  # Ajusta el espaciado
plt.show()  # Muestra los gráficos
```
